```python
#TITLE



import matplotlib.pyplot as plt
import numpy as np
x =np.array([0,6,10,40,33])
x1 =np.array([0,50,60,70,80])
plt.plot(x,x1)
plt.title('x and y dia')
plt.show()
```

```python
#TITLE,label



import matplotlib.pyplot as plt
import numpy as np
x =np.array([0,6,10,40,33])
x1 =np.array([0,50,60,70,80])
plt.plot(x,x1)
plt.title('x and y dia')
plt.ylabel("this is y axis")
plt.xlabel("this is x axis")
plt.show()
```

```python
#TITLE,label,grids



import matplotlib.pyplot as plt
import numpy as np
x =np.array([0,6,10,40,33])
x1 =np.array([0,50,60,70,80])
plt.plot(x,x1)
plt.title('x and y dia')
plt.ylabel("this is y axis")
plt.xlabel("this is x axis")
plt.grid()
plt.show()
```

```python
#grid for particular axis


#TITLE,label,grids



import matplotlib.pyplot as plt
import numpy as np
x =np.array([0,6,10,40,33])
x1 =np.array([0,50,60,70,80])
plt.plot(x,x1)
plt.title('x and y dia')
plt.ylabel("this is y axis")
plt.xlabel("this is x axis")
plt.grid(axis ="x",color ="y")
plt.show()
```

```python
#subplot

import matplotlib.pyplot as plt
import numpy as np
x =np.array([0,1,2,3,4])
x1 =np.array([0,50,60,70,80])
plt.subplot(2,1,1)
plt.plot(x,x1)


plt.title('x and y dia')
plt.ylabel("this is y axis")
plt.xlabel("this is x axis")
plt.grid(axis ="x",color ="y")



#plot 2
x =np.array([0,10,20,30,40])
x1 =np.array([0,500,600,700,800])
plt.subplot(2,1,2)
plt.plot(x,x1)






plt.title('x and y dia')
plt.ylabel("this is y axis")
plt.xlabel("this is x axis")
plt.grid(axis ="x",color ="y")
plt.show()

```

```python
#supertitle

import matplotlib.pyplot as plt
import numpy as np
x =np.array([0,1,2,3,4])
x1 =np.array([0,50,60,70,80])
plt.subplot(2,1,1)
plt.plot(x,x1)


plt.title('x and y dia')
plt.ylabel("this is y axis")
plt.xlabel("this is x axis")
plt.grid(axis ="x",color ="y")



#plot 2
x =np.array([0,10,20,30,40])
x1 =np.array([0,500,600,700,800])
plt.subplot(2,1,2)
plt.plot(x,x1)
plt.suptitle("this is a worked demo")






plt.title('x and y dia')
plt.ylabel("this is y axis")
plt.xlabel("this is x axis")
plt.grid(axis ="x",color ="y")
plt.show()

```

```python
#Bar chart
import matplotlib.pyplot as plt
import numpy as np
x =np.array(["a","b","c","d","e"])
y =np.array([0,50,60,70,80])
plt.bar(x,y)
plt.show()
```

```python
#for horizontal

import matplotlib.pyplot as plt
import numpy as np
x =np.array(["a","b","c","d","e"])
y =np.array([0,50,60,70,80])
plt.barh(x,y)
plt.show()
```

```python
#color

#Bar chart
import matplotlib.pyplot as plt
import numpy as np
x =np.array(["a","b","c","d","e"])
y =np.array([0,50,60,70,80])
plt.bar(x,y,color ='y')
plt.show()
```

```python
#histogram
import matplotlib.pyplot as plt
import numpy as np
x =np.random.normal(160,10,250)
plt.hist(x,color='g')
plt.show()
```

```python
#piechart

import matplotlib.pyplot as plt
import numpy as np
x =np.random.normal(160,10,250)
plt.pie(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x =np.array([160,10,250,80,100])
mylabel =["ravi","partha","mathan","abi","sakthi"]
plt.pie(x,labels=mylabel)
plt.show()
```

```python
#with start angle

import matplotlib.pyplot as plt
import numpy as np
x =np.array([160,10,250,80,100])
mylabel =["ravi","partha","mathan","abi","sakthi"]
plt.pie(x,labels=mylabel,startangle =90)
plt.show()
```

```python
#highlight

import matplotlib.pyplot as plt
import numpy as np
x =np.array([160,10,250,80,100])
mylabel =["ravi","partha","mathan","abi","sakthi"]
seperate =[0.2,0,0,0,0]
plt.pie(x,labels=mylabel,explode=seperate)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x =np.array([160,10,250,80,100])
mylabel =["ravi","partha","mathan","abi","sakthi"]
seperate =[0.2,0.1,0.3,0,0]
mycolors =["pink","red","green","black","#4CFF33"]
plt.pie(x,labels=mylabel,explode=seperate,colors=mycolors)
plt.show()
```
